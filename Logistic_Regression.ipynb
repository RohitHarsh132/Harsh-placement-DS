 Logistic Regression is a Classification Algorithm used to predict Binary Outcomes(Yes No , 0 or 1) based on input Data.

Problem: You want to predict if someone will buy Ice-Creame (Yes/No) depending on the temperature.

At 25°C → Maybe No ❌

At 35°C → Maybe Yes ✅

You can’t use Linear Regression here, because the outcome is not a number — it’s a category.

Sigmoid Function is a Mathematical Function that takes any number and convert into a value between 0 and 1.

sigmoid(z)=1/1+e^-z


```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
Temp=np.array([20,25,30,35,40]).reshape(-1,1)
Buy=np.array([0,0,1,1,1])
```

```python
model=LogisticRegression()
model.fit(Temp,Buy)
```

```python
test_temp=[[28]]
prob=model.predict_proba(test_temp)
print(f"Probability of buying at 28°C: {prob[0][1]:.2f}")
```

```python
# Step 5: Visualize the Logistic Curve
temp_range = np.linspace(15, 45, 300).reshape(-1, 1)
probabilities = model.predict_proba(temp_range)[:, 1]

plt.plot(temp_range, probabilities, color='red', label='Probability Curve')
plt.scatter(Temp, Buy, color='blue', label='Actual')
plt.xlabel('Temperature (°C)')
plt.ylabel('Probability of Buying Ice-cream')
plt.title('Logistic Regression: Buy or Not')
plt.legend()
plt.grid(True)
plt.show()
```
