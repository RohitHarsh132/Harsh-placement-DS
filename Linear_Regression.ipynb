Problem Statement
: You are Running Ice creame shop. You want to predict how many ice-creame You will sell based on the Temparature.

Goal:Build a linear regression model to learn the relationship:

# Sales = m × Temperature + c

# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
temp=np.array([25,30,35,40,45]).reshape(-1,1) # convert 1D array to 2D coloumn vector
sales=np.array([30,40,50,60,70])
```

```python
model= LinearRegression()
model.fit(temp, sales)
```

```python
temp_input = [[33]]
predicted_sales=model.predict(temp_input)
print(f"Predicted lemonade sales at 33°C: {predicted_sales[0]:.2f}")
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print(f"Learned Equation: Sales = {slope:.2f} × Temperature + {intercept:.2f}")
```

```python
# Step 6: Visualize the Data and the Regression Line
plt.scatter(temp, sales, color='blue', label='Actual Sales')  # Original data
plt.plot(temp, model.predict(temp), color='red', label='Regression Line')  # Line of best fit
plt.xlabel('Temperature (°C)')
plt.ylabel('Icew-Creame Sold')
plt.title('Ice-creame Sales vs Temperature')
plt.legend()
plt.grid(True)
plt.show()
```
